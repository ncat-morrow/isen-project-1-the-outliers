## Histogram Plotting
Plot a histogram for each of the three features of the dataset. Be sure to include proper titles and axis labels on each of your plots.

````python
import matplotlib.pyplot as plt
import pandas as pd # Import pandas

df = data # Load the data into a DataFrame called 'df'

# Plot histogram for Heating Water Flow Rate
plt.hist(df['Heating Water Flow Rate'], bins=20)
plt.title('Histogram of Heating Water Flow Rate')
plt.xlabel('Heating Water Flow Rate')
plt.ylabel('Frequency')
plt.show()

# Plot histogram for Cooling Water Flow Rate
plt.hist(df['Cooling Water Flow Rate'], bins=20)
plt.title('Histogram of Cooling Water Flow Rate')
plt.xlabel('Cooling Water Flow Rate')
plt.ylabel('Frequency')
plt.show()

# Plot histogram for TotalEnergy
plt.hist(df['TotalEnergy'], bins=20)
plt.title('Histogram of TotalEnergy')
plt.xlabel('TotalEnergy')
plt.ylabel('Frequency')
plt.show()
````
